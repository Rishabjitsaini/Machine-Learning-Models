```python
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
##stage 1 Initialization stage
##Creating dataFrame Structure
```

```python
iris=datasets.load_iris()
```

```python
df=pd.DataFrame({'x':iris.data[:,0],'y':iris.data[:,1],'cluster':iris.target})
df
```

```python
centroids={}
for i in range(3):
    result_list=[]
    result_list.append(df.loc[df['cluster']==i]['x'].mean())
    result_list.append(df.loc[df['cluster']==i]['y'].mean())
    
    centroids[i]=result_list
centroids
```

```python
##plotting points
```

```python
fig=plt.figure(figsize=(5,5))
plt.scatter(df['x'],df['y'],c=iris.target)
plt.xlabel('Speal length',fontsize=20)
plt.ylabel("Sepal Width",fontsize=20)
```

```python
##defining colors and ploting centroids for all clusters
```

```python
colmap={0:'r',1:'g',2:'b'}
for i in range(3):
    plt.scatter(centroids[i][0],centroids[i][1],color=colmap[i])
plt.show()
```

```python
##showing complete combine graph
```

```python
fig=plt.figure(figsize=(5,5))
plt.scatter(df['x'],df['y'],c=iris.target,alpha=0.3)
colmap={0:'r',1:'g',2:'b'}
col=[0,1]
for i in centroids.keys():
     plt.scatter(centroids[i][0],centroids[i][1],color=colmap[i],edgecolor='k')
plt.show()
    
```

```python
##stage2 assignment fuction
##calculating distance and updating data frame
```

```python
def assignment(df,centroids):
    for i in range(3):
        #sqrt((x1-x2)^2+(y1-y2)^2)
        df['distance_from_{}'.format(i)]=(
            np.sqrt(
                (df['x']-centroids[i][0])**2
                +(df['y']-centroids[i][1])**2
            )
        )
    centroid_distance_cols=['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest']=df.loc[:,centroid_distance_cols].idxmin(axis=1)
    df['closest']=df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color']=df['closest'].map(lambda x:colmap[x])
    return df
```

```python
df=assignment(df,centroids)
df
```

```python
fig=plt.figure(figsize=(6,8))
plt.scatter(df['x'],df['y'],color=df['color'],alpha=0.3)
for i in centroids.keys():
     plt.scatter(*centroids[i],color=colmap[i],edgecolor='k')
plt.show()
```

```python
##stage 3:update stage
#update fuction: updating the centroids
```

```python
def update(k):
    for i in range(3):
        centroids[i][0]=np.mean(df[df['closest']==i]['x'])
        centroids[i][1]=np.mean(df[df['closest']==i]['y'])
    return k
centroids=update(centroids)
centroids
```

```python
#visualising the graph with updated centroids
```

```python
fig=plt.figure(figsize=(6,8))
plt.scatter(df['x'],df['y'],color=df['color'],alpha=0.3)
for i in centroids.keys():
     plt.scatter(*centroids[i],color=colmap[i],edgecolor='k')
plt.show()
```

```python
#repeat assignment stage to resign the points
```

```python
df=assignment(df,centroids)
```

```python
#visualising the update graph
```

```python
fig=plt.figure(figsize=(5,5))
plt.scatter(df['x'],df['y'],color=df['color'],alpha=0.3)
for i in centroids.keys():
     plt.scatter(centroids[i][0],centroids[i][1],color=colmap[i],edgecolor='k')
plt.show()
```

```python
##continuing until all assigned cluster dont change any more
```

```python
while True:
    closest_centroids=df['closest'].copy(deep=True)
    centroids=update(centroids)
    df=assignment(df,centroids)
    if closest_centroids.equals(df['closest']):
        break
```

```python
#final result
```

```python
fig=plt.figure(figsize=(5,5))
plt.scatter(df['x'],df['y'],color=df['color'],alpha=0.3)
for i in centroids.keys():
     plt.scatter(centroids[i][0],centroids[i][1],color=colmap[i],edgecolor='k')
plt.show()
```
