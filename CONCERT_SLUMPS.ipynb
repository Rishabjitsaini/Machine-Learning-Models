```python
x = df.drop(columns ="Compressive Strength (28-day)(Mpa)") 
y = df["Compressive Strength (28-day)(Mpa)"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state =42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_x_train=scaler.fit_transform(x_train)
```

```python
scaled_x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.svm import SVR, LinearSVR
```

```python
base_model=SVR()
```

```python
base_model.fit(scaled_x_train, y_train)
```

```python
base_preds=base_model.predict(scaled_x_test)
base_preds
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test,base_preds)
```

```python
np.sqrt(mean_absolute_error(y_test,base_preds))
```

```python
y_test.mean()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.001,0.01,0.1,0.5,1],'kernel':['linear','rbf','poly'],'gamma':['scale','auto'],'degree':[2,3,4],'epsilon':[0,0.01,0.1,0.5,1,2]}
```

```python
param_grid
```

```python
svr=SVR()
```

```python
grid_model=GridSearchCV(svr,param_grid)
```

```python
grid_model.fit(scaled_x_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_preds=grid_model.predict(scaled_x_test)
```

```python
mean_absolute_error(y_test,grid_preds)
```

```python
np.sqrt(mean_squared_error(y_test,grid_preds))
```
